```python
import pandas as pd
import numpy as np
import patsy as pt
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model, metrics, model_selection, preprocessing
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Clean up the data obtained from the .rda file
wage_dat = pd.read_csv('wage.csv')
wage_dat = wage_dat.drop(wage_dat.columns[0], axis=1)
wage_dat.head()
```

```python
# SKLearn's LinearRegression class does not calculate the the pointwise standard error
# - One can achieve this as follows:
# - This was helpful - https://stats.stackexchange.com/questions/44838/how-are-the-standard-errors-of-coefficients-calculated-in-a-regression
# - For future reference, code to calculate p-values by hand - https://www.quora.com/In-scikit-learn-how-can-you-obtain-the-standard-errors-of-regression-coefficients

def gen_coeff_covar_matrix(true, preds, d_matrix):
    '''calculate covariance matrix for regression coefficients'''
    mse = metrics.mean_squared_error(true, preds)
    return mse * np.linalg.inv(d_matrix.T @ d_matrix)
    
def gen_pointwise_se_curves(cov_matrix, preds, d_matrix):
    '''calculate pointwise standard error: see last paragraph and footnote on p267 ISL'''
    var_f_hat = np.diagonal((d_matrix @ cov) @ d_matrix.T)
    return pd.DataFrame({'error_pos' : preds + 2 * np.sqrt(var_f_hat), \
                         'error_neg' : preds - 2 * np.sqrt(var_f_hat)})
```

```python
# Polynomial Regression with SKLearn

# could also be achieved with `preprocessing.PolynomialFeatures`..
f = ' + '.join(['np.power(age, {})'.format(i) for i in range(1,5)])
d_matrix = pt.dmatrix(f, wage_dat)
model = linear_model.LinearRegression(fit_intercept=False).fit(d_matrix, wage_dat['wage'])
preds = model.predict(d_matrix)

# plot scatter
_, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(x='age', y='wage', data=wage_dat, ax=ax, color='tab:gray')

# plot line
reg_df = pd.DataFrame({'age' : wage_dat['age'], 'wage' : preds})
sns.lineplot(x='age', y='wage', data=reg_df, color='tab:blue')

# plot error
cov = gen_coeff_covar_matrix(wage_dat['wage'], preds, d_matrix)
err_curves = gen_pointwise_se_curves(cov, preds, d_matrix)
err_curves['age'] = wage_dat['age']
sns.lineplot(x='age', y='error_pos', color='y', data = err_curves)
sns.lineplot(x='age', y='error_neg', color='y', data = err_curves)
```

```python
# Selecting the order of the polynomial with CV

up_to_order = 5
df = pd.DataFrame()
for train_idx, test_idx in model_selection.KFold(n_splits=10).split(wage_dat):
    train, test = wage_dat.iloc[train_idx], wage_dat.iloc[test_idx]
    errors = []
    for order in range(1, up_to_order + 1):
        f = ' + '.join(['np.power(age, {})'.format(i) for i in range(1, order + 1)])
        train_d_matrix = pt.dmatrix(f, train)
        model = linear_model.LinearRegression(fit_intercept=False).fit(train_d_matrix, train['wage'])
        test_d_matrix = pt.dmatrix(f, test)
        preds = model.predict(test_d_matrix)
        errors.append(np.sqrt(metrics.mean_squared_error(test['wage'], preds)))
    df = df.append(pd.Series(errors), ignore_index=True)

df.columns = df.columns + 1
df.index = df.index + 1
# render dataframe with rmse for each order polynomial for each split
df
```

```python
# render sorted mean rmses
means = df.mean()
sns.barplot(x='order', y='rmse', data=pd.DataFrame({'order' : means.index, 'rmse' : means}))
```

```python
# Constant Step Function

cuts = pd.cut(wage_dat['age'], 4, labels=['a1', 'a2', 'a3', 'a4'])
df = pd.get_dummies(cuts)
df = pd.concat([pd.DataFrame(wage_dat['wage']), df], axis=1)
d_matrix = pt.dmatrix('a1 + a2 + a3 + a4', df)

model = linear_model.LinearRegression(fit_intercept=False).fit(d_matrix, wage_dat['wage'])
preds = model.predict(d_matrix)


# plot scatter
_, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(x='age', y='wage', data=wage_dat, ax=ax, color='tab:gray')

# plot line
reg_df = pd.DataFrame({'age' : wage_dat['age'], 'wage' : preds})
sns.lineplot(x='age', y='wage', data=reg_df, color='tab:blue')

# plot error
cov = gen_coeff_covar_matrix(wage_dat['wage'], preds, d_matrix)
err_curves = gen_pointwise_se_curves(cov, preds, d_matrix)
err_curves['age'] = wage_dat['age']
sns.lineplot(x='age', y='error_pos', color='y', data = err_curves)
sns.lineplot(x='age', y='error_neg', color='y', data = err_curves)
```

```python
# Splines

x = np.linspace(0., 1., 100)
y = pt.dmatrix("bs(x, df=6, degree=3, include_intercept=True) - 1", {"x": x})
b = np.array([1.3, 0.6, 0.9, 0.4, 1.6, 0.7])
# Plot B-spline basis functions (colored curves) each multiplied by its coeff
plt.plot(x, y*b)
# Plot the spline itself (sum of the basis functions, thick black curve)
plt.plot(x, np.dot(y, b), color='k', linewidth=3)
```

```python
# Spline

d_matrix = pt.dmatrix('bs(age, df=7, degree=3, include_intercept=True)', wage_dat)
model = linear_model.LinearRegression(fit_intercept=False).fit(d_matrix, wage_dat['wage'])
preds = model.predict(d_matrix)

# plot scatter
_, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(x='age', y='wage', data=wage_dat, ax=ax, color='tab:gray')

# plot line
reg_df = pd.DataFrame({'age' : wage_dat['age'], 'wage' : preds})
sns.lineplot(x='age', y='wage', data=reg_df, color='tab:blue')

# plot error
cov = gen_coeff_covar_matrix(wage_dat['wage'], preds, d_matrix)
err_curves = gen_pointwise_se_curves(cov, preds, d_matrix)
err_curves['age'] = wage_dat['age']
sns.lineplot(x='age', y='error_pos', color='tab:blue', data = err_curves)
sns.lineplot(x='age', y='error_neg', color='tab:blue', data = err_curves)
ax.lines[1].set_linestyle("--")
ax.lines[2].set_linestyle("--")


# Natural Spline

d_matrix = pt.dmatrix('cr(age, df=4)', wage_dat)
model = linear_model.LinearRegression(fit_intercept=False).fit(d_matrix, wage_dat['wage'])
preds = model.predict(d_matrix)

# plot line
reg_df = pd.DataFrame({'age' : wage_dat['age'], 'wage' : preds})
sns.lineplot(x='age', y='wage', data=reg_df, color='tab:red')

cov = gen_coeff_covar_matrix(wage_dat['wage'], preds, d_matrix)
err_curves = gen_pointwise_se_curves(cov, preds, d_matrix)
err_curves['age'] = wage_dat['age']
sns.lineplot(x='age', y='error_pos', color='tab:red', data = err_curves)
sns.lineplot(x='age', y='error_neg', color='tab:red', data = err_curves)
ax.lines[4].set_linestyle("--")
ax.lines[5].set_linestyle("--")
```
